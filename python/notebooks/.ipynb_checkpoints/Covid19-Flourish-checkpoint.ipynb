# DATA FOR FLOURISH @ https://app.flourish.studio/@psycho.presley

```python
file_dir = r'C:\Users\user\Documents\GitHub\COVID-19\consolidated_data\country_report.json'
save_file = r'C:\Users\user\Documents\GitHub\COVID-19\consolidated_data\flourish.csv'
```

```python
df = pd.read_json(file_dir)
```

```python
list_of_variables = ['Active','Confirmed','Deaths','Recovered']
initial_date = '2020-03-06'

def generate_flourish(df,parameters='Confirmed',initial_date):
    '''
    With this function it is possible to generate
    the dataset as used by Florish Studio @
    https://app.flourish.studio/@psycho.presley



    '''
    import pandas as pd
    import os

    countries = df['Country/Region'].unique()
    
    for item in parameters:
        columns = ['Country/Region','Date',item]
        flourish = pd.DataFrame()

        for country in countries:
            df_aux = df[columns].loc[df['Country/Region'] == country]
            df_aux = df_aux.pivot(index='Country/Region',columns='Date', values='Active')

            flourish = pd.concat([flourish,df_aux]).interpolate(method='linear',limit=3)
            flourish.fillna(method='bfill',inplace=True)

        flourish.loc[:,initial_date:].to_csv(save_file)
```
